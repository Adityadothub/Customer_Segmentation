```python
# STEP 1: Install and import required packages
!pip install gdown

import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import gdown
```

```python
# STEP 2: Download the file from Google Drive
# Extracted from your link: https://drive.google.com/file/d/195kefK8Ff9BGIhKLQcipWdMxmnZDwvOT/view
file_id = "195kefK8Ff9BGIhKLQcipWdMxmnZDwvOT"
url = f"https://drive.google.com/uc?id={file_id}"

# Download and save as a CSV
gdown.download(url, "customers.csv", quiet=False)

# Load the data
data = pd.read_csv("customers.csv")
data.head()
```

```python
# STEP 3: Create SQLite DB and load data
conn = sqlite3.connect('customer_segmentation.db')
data.to_sql('customers', conn, if_exists='replace', index=False)

# View some data
pd.read_sql('SELECT * FROM customers LIMIT 5', conn)
```

```python
# STEP 4: Extract relevant columns using SQL
query = """
SELECT Age, `Annual Income (k$)`, `Spending Score (1-100)`
FROM customers
"""
df = pd.read_sql(query, conn)
df.columns = ['Age', 'Income', 'Score']  # Rename for easier use
df = df.dropna()
df.head()
```

```python
# STEP 5: KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)
df.head()
```

```python
# STEP 6: Visualize clusters
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Income', y='Score', hue='Cluster', palette='Set2')
plt.title('Customer Segmentation')
plt.show()
```
